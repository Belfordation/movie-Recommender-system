# Finding Similar Movies
we'll start by loading up the MovieLens dataset.using pandas we can very quickly load th rows of the rate.csv & movie.csv files that we care about and merge them toghether so we can work with movie names instead of ID's(in a real production job you'd stick with ID's and worry about names at the display layer to make things more efficient.but this lets us understand what's going on better for now.)

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
r_cols=['userId','movieId','rating']
ratings=pd.read_csv('F:\DATA SCIENCE\DATAZ\Recommender system\RATE.csv',names=r_cols,usecols=range(3))

m_cols=['movieId','title']
movies=pd.read_csv('F:\DATA SCIENCE\DATAZ\Recommender system\MOVIE.csv',names=m_cols,usecols=range(2))

ratings=pd.merge(movies,ratings,how='inner',on='movieId')
```

```python
ratings.head()
```

Now the amazing pivot_table function on a dataframe will construct a user/movie rating matrix.Note how NAN indicates missing data-movies that specific users didn't rate.

```python
movieRatings=ratings.pivot_table(index=('userId'),columns=['title'],values='rating')
movieRatings.head()
```

let's extract a series of users who rated for toy story:

```python
ToyStoryRatings=movieRatings['Toy Story (1995)'] 
ToyStoryRatings.head()
```

pandas' corrwith function makes it really easy to compute the pairwise correlation of toy story's vector of user rating with every other movie!after that,we'll drop any results that have no data,and construct a new Dataframe of movies and their correlation score(similarity)to toy story: 

```python
similarMovies=movieRatings.corrwith(ToyStoryRatings)
similarMovies=similarMovies.dropna()
df=pd.DataFrame(similarMovies)
df.head(10)
```

let's sort the results by similarity score,and we shoulld have the movies most similar to toy story.except...we don't.these results make no sense at all this is why it's important to know your data.

clearly we missed somthing important


```python
similarMovies.sort_values(ascending=False)
```

our results are probably gettig messed up by movies that only have been viewd by a handful of people who also happened to like star wars.so we need to get rid of movies that were only watched by a few people that are producing spurious results.let's construct a new DataFrame that counts up how many ratings exist for each movie,and alsothe average rating while we're at it-could also come in handy later

```python
movieStats=ratings.groupby('title').agg({'rating':[np.size,np.mean]})
movieStats.tail()                                         
```

let's get rid of movies rated by fewer than 100 people,and check the top-rated ones that are left:

```python
popularMovies=movieStats['rating']['size']>=100
movieStats[popularMovies].sort_values([('rating','mean')],ascending=False)[:15] #a koja fahmid in ascending bara meane
```


100 might still be too low,but these results look pretty good as far as "well rated movies that people have heard of".let's join this Data with our original set of similar movies to TOY STORY:

```python
 df=movieStats[popularMovies].join(pd.DataFrame(similarMovies,columns=['similarity']))
```

```python
df.head()
```

And,sort these new results by similarity score.that's more like it!

```python
df.sort_values(['similarity'],ascending=False)[:15]
```
